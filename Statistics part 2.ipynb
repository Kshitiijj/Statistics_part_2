```python
# 1. What is hypothesis testing in statistics?
"""
Hypothesis testing is a statistical method used to determine if there is enough evidence in a sample to infer a condition for
the entire population. It involves formulating a null hypothesis (H₀) and an alternative hypothesis (H₁) and using statistical tests to decide whether to reject H₀.
"""

# 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
"""
The null hypothesis (H₀) represents a statement of no effect or no difference. It assumes that observed results are due to chance. 
The alternative hypothesis (H₁) suggests that there is a significant effect or difference that is not due to random chance.
"""

# 3. What is the significance level in hypothesis testing, and why is it important?
"""
The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. 
Common values are 0.05 or 0.01. It determines the threshold for statistical significance and helps control the Type I error rate.
"""

# 4. What does a P-value represent in hypothesis testing?
"""
The P-value measures the probability of obtaining the observed data (or more extreme results) assuming that the null hypothesis is true. 
A smaller P-value indicates stronger evidence against H₀.
"""

# 5. How do you interpret the P-value in hypothesis testing?
"""
If the P-value is less than the significance level (α), we reject H₀, suggesting evidence in favor of H₁. 
If P-value > α, we fail to reject H₀, meaning there is not enough evidence to support H₁.
"""

# 6. What are Type 1 and Type 2 errors in hypothesis testing?
"""
Type I error occurs when we wrongly reject a true null hypothesis (false positive). 
Type II error happens when we fail to reject a false null hypothesis (false negative).
"""

# 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
"""
A one-tailed test checks for an effect in only one direction (e.g., greater than). 
A two-tailed test checks for an effect in both directions (e.g., not equal to).
"""

# 8. What is the Z-test, and when is it used in hypothesis testing?
"""
A Z-test is used to determine if there is a significant difference between sample and population means when the population variance
is known and the sample size is large (n > 30).
"""

# 9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
"""
Z-score = (X̄ - μ) / (σ / √n), where X̄ is the sample mean, μ is the population mean, σ is the population standard deviation, 
and n is the sample size. It represents how many standard deviations a data point is from the population mean.
"""

# 10. What is the T-distribution, and when should it be used instead of the normal distribution?
"""
The T-distribution is used when the population standard deviation is unknown and the sample size is small (n < 30). 
It is similar to the normal distribution but has heavier tails.
"""

# 11. What is the difference between a Z-test and a T-test?
"""
A Z-test is used when the population variance is known and for large samples, while a T-test is used when the population variance 
is unknown and for small samples.
"""

# 12. What is the T-test, and how is it used in hypothesis testing?
"""
A T-test compares means between two groups to determine if they are significantly different from each other. 
It is commonly used in small sample studies.
"""

# 13. What is the relationship between Z-test and T-test in hypothesis testing?
"""
Both tests assess differences between means, but a Z-test is used for large samples with known variance, 
while a T-test is used for small samples with unknown variance.
"""

# 14. What is a confidence interval, and how is it used to interpret statistical results?
"""
A confidence interval is a range of values derived from sample data that is likely to contain the population parameter. 
A 95% confidence interval means we are 95% confident that the population parameter lies within the interval.
"""

# 15. What is the margin of error, and how does it affect the confidence interval?
"""
The margin of error represents the maximum expected difference between the sample statistic and the true population parameter. A
larger margin of error means a wider confidence interval.
"""

# 16. How is Bayes' Theorem used in statistics, and what is its significance?
"""
Bayes' Theorem describes how to update the probability of a hypothesis based on new evidence. 
It is used in probability theory, machine learning, and Bayesian statistics.
"""

# 17. What is the Chi-square distribution, and when is it used?
"""
The Chi-square distribution is used for hypothesis testing on categorical data, 
including goodness-of-fit tests and tests for independence in contingency tables.
"""

# 18. What is the Chi-square goodness of fit test, and how is it applied?
"""
The Chi-square goodness-of-fit test determines whether an observed frequency distribution matches an expected distribution. 
It is applied in categorical data analysis.
"""

# 19. What is the F-distribution, and when is it used in hypothesis testing?
"""
The F-distribution is used in ANOVA and regression analysis to compare variances between groups. 
It is right-skewed and depends on degrees of freedom.
"""

# 20. What is an ANOVA test, and what are its assumptions?
"""
ANOVA (Analysis of Variance) is used to compare means across multiple groups. 
Assumptions include normality, homogeneity of variances, and independent observations.
"""

# 21. What are the different types of ANOVA tests?
"""
Types include one-way ANOVA (one independent variable), two-way ANOVA (two independent variables), and repeated measures 
ANOVA (same subjects tested multiple times).
"""

# 22. What is the F-test, and how does it relate to hypothesis testing?
"""
The F-test compares variances to test if two or more groups have equal variances. 
It is used in ANOVA and regression analysis to determine statistical significance.
"""
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
from statsmodels.stats.weightstats import ztest
sample_data = np.array([50, 52, 48, 51, 49, 47, 50, 53, 54, 46])
pop_mean = 50
z_stat, p_value = ztest(sample_data, value=pop_mean)
print("Z-statistic:", z_stat, "P-value:", p_value)
```

```python
# 2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.
data = np.random.normal(loc=100, scale=15, size=50)
z_stat, p_value = ztest(data, value=100)
print("Simulated Z-statistic:", z_stat, "P-value:", p_value)


```

```python
# 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
def one_sample_z_test(data, pop_mean):
    z_stat, p_value = ztest(data, value=pop_mean)
    return z_stat, p_value

sample = np.random.normal(55, 5, 30)
z_stat, p_value = one_sample_z_test(sample, 50)
print("One-Sample Z-Test: Z-stat:", z_stat, "P-value:", p_value)
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)
plt.plot(x, y, label="Normal Distribution")
plt.axvline(-1.96, color='red', linestyle='dashed', label="Critical Value (-1.96)")
plt.axvline(1.96, color='red', linestyle='dashed', label="Critical Value (1.96)")
plt.fill_between(x, y, where=(x < -1.96) | (x > 1.96), color='red', alpha=0.3)
plt.legend()
plt.title("Two-Tailed Z-Test Decision Region")
plt.show()

```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
def visualize_errors():
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)
    plt.plot(x, y, label="H₀ Distribution", color="blue")
    plt.axvline(1.96, color='red', linestyle='dashed', label="Type 1 Error")
    plt.fill_between(x, y, where=(x > 1.96), color='red', alpha=0.3)
    plt.plot(x, stats.norm.pdf(x, loc=2), label="H₁ Distribution", color="green")
    plt.fill_between(x, stats.norm.pdf(x, loc=2), where=(x < 1.96), color='green', alpha=0.3)
    plt.legend()
    plt.title("Type 1 and Type 2 Errors")
    plt.show()

visualize_errors()

```

```python
# 6. Write a Python program to perform an independent T-test and interpret the results.
group1 = np.random.normal(100, 10, 30)
group2 = np.random.normal(105, 10, 30)
t_stat, p_value = stats.ttest_ind(group1, group2)
print("Independent T-Test: T-stat:", t_stat, "P-value:", p_value)

```

```python
# 7. Perform a paired sample T-test using Python and visualize the comparison results.
before = np.random.normal(100, 5, 30)
after = before + np.random.normal(0, 2, 30)
t_stat, p_value = stats.ttest_rel(before, after)
print("Paired Sample T-Test: T-stat:", t_stat, "P-value:", p_value)

plt.hist(before, alpha=0.5, label="Before")
plt.hist(after, alpha=0.5, label="After")
plt.legend()
plt.title("Paired Sample Data Distribution")
plt.show()

```

```python
# 8. Simulate data and perform both Z-test and T-test, then compare the results using Python.
simulated_data = np.random.normal(100, 10, 40)
z_stat, z_p_value = ztest(simulated_data, value=100)
t_stat, t_p_value = stats.ttest_1samp(simulated_data, 100)
print(f"Z-Test: Z-stat={z_stat}, P-value={z_p_value}")
print(f"T-Test: T-stat={t_stat}, P-value={t_p_value}")

```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    std_err = stats.sem(data)
    margin = std_err * stats.t.ppf((1 + confidence) / 2., len(data) - 1)
    return mean - margin, mean + margin

sample_data = np.random.normal(100, 10, 30)
ci = confidence_interval(sample_data)
print("95% Confidence Interval:", ci)
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
def margin_of_error(data, confidence=0.95):
    std_err = stats.sem(data)
    margin = std_err * stats.t.ppf((1 + confidence) / 2., len(data) - 1)
    return margin

sample_data = np.random.normal(100, 10, 30)
print("Margin of Error:", margin_of_error(sample_data))

```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayes_theorem(prior, likelihood, evidence):
    return (likelihood * prior) / evidence

prior = 0.5
likelihood = 0.8
evidence = 0.6
print("Bayesian Probability:", bayes_theorem(prior, likelihood, evidence))

```

```python
# 12. Perform a Chi-square test for independence between two categorical variables in Python.
from scipy.stats import chi2_contingency

data = np.array([[10, 20, 30], [6, 9, 17]])
chi2, p, dof, expected = chi2_contingency(data)
print("Chi-Square Test: Chi2 =", chi2, "P-value =", p)

```

```python
# 13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
def expected_frequencies(observed):
    _, _, _, expected = chi2_contingency(observed)
    return expected

observed_data = np.array([[15, 25], [10, 20]])
print("Expected Frequencies:\n", expected_frequencies(observed_data))

```

```python
# 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.
def check_anova_assumptions(data_groups):
    for group in data_groups:
        print("Shapiro-Wilk Normality Test:", stats.shapiro(group))
    print("Levene's Test for Equal Variance:", stats.levene(*data_groups))

check_anova_assumptions([group1, group2, group3])

```

```python
# 20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.
import statsmodels.api as sm
from statsmodels.formula.api import ols

df = pd.DataFrame({
    'Factor1': np.tile(['A', 'B'], 15),
    'Factor2': np.repeat(['X', 'Y', 'Z'], 10),
    'Values': np.random.normal(100, 10, 30)
})
model = ols('Values ~ C(Factor1) + C(Factor2) + C(Factor1):C(Factor2)', data=df).fit()
anova_results = sm.stats.anova_lm(model, typ=2)
print("Two-Way ANOVA Results:\n", anova_results)

```

```python
# 24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
variance = np.var(data_simulated, ddof=1)
chi2_stat = (len(data_simulated) - 1) * variance / 100
p_value = stats.chi2.sf(chi2_stat, df=len(data_simulated)-1)
print("Hypothesis Test for Population Variance: Chi2 =", chi2_stat, "P-value =", p_value)

```

```python
# 25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.
success1, total1 = 50, 100
success2, total2 = 40, 90
p1, p2 = success1/total1, success2/total2
p_combined = (success1 + success2) / (total1 + total2)
z_stat = (p1 - p2) / np.sqrt(p_combined * (1 - p_combined) * (1/total1 + 1/total2))
p_value = stats.norm.sf(abs(z_stat)) * 2
print("Z-Test for Proportions: Z-stat =", z_stat, "P-value =", p_value)

```
